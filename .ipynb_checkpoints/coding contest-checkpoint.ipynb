```python
class Solution():
    
    def lengthOfLongestSubstring(self, s: str) -> int:
        l = len(s)
        if l == 1:
            return 1
        m, start, dic = 0, 0, {}
        for i, v in enumerate(s):
            if v in dic and dic[v] >= start :
                start = dic[v] + 1
            else:
                m = max(i - start + 1, m)
            dic[v] = i
        return m
```

```python
from typing import List
class Solution:
    def groupThePeople(self, groupSizes: List[int]) -> List[List[int]]:
        s = list(set(groupSizes))
        l = len(s)
        lst = [[] for _ in range(l)]
        for n, m in enumerate(s):
            for i, num in enumerate(groupSizes):
                if num == m:
                    lst[n].append(i)
        ret = []
        for item, ls in zip(lst, s):
            q = len(item) // ls
            for mod in range(q):
                ret.append(item[mod * ls : (mod + 1) * ls])
        return ret
```

```python
from math import ceil

class Solution:
    def smallestDivisor(self, nums: List[int], threshold: int) -> int:
        m = 0
        s = sum(nums)
        l = len(nums)
        for n in range(s // threshold, s - threshold):
            if n == 0:
                continue
            qlst = [ceil(i/n) for i in nums]
            sq = sum(qlst)
            if sq <= threshold:
                return n
            n += 1
```

```python
import numpy as np
H, W = map(int, input().split())
lst1 = []
lst2 = []
for _ in range(H):
    lst1.append([int(i) for i in input().split()])
for _ in range(H):
    lst2.append([int(i) for i in input().split()])
arr1, arr2 = np.array(lst1), np.array(lst2)
arr = np.abs(arr1 - arr2)

d = [[0] * W for _ in range(H)]
m = 80 * (H + W)  # 400
d[0][0] = 1 << m  # 最初の「位置」
for h in range(H):
    for w in range(W):
        x = arr[h][w]  # 現在地の偏りを取る
        if h > 0:
            print(h,w)
            d[h][w] |= d[h - 1][w]  # 前回のすべて可能な位置を継承する
        if w > 0:
            d[h][w] |= d[h][w - 1]  # 前回のすべて可能な位置を継承する
        d[h][w] = d[h][w] << x | d[h][w] >> x
'''
今回可能な移動を記録する。
例えば前回の記録は、
00010(0,ここは最初の位置)01000で、
xは2の時、偏りは2と-2の二つ可能性がある。
ならば今回の記録は、
01000(1,ここは最初の位置)00000 | 00000(1,ここは最初の位置)00010
= 01000(1,ここは最初の位置)00010
になる。
これは、「偏りが4,0,-4の３つ可能性がある」を表示する。
全部やった後、d[H - 1][W - 1]がすべて可能な偏りを記録したことがわかる。
'''
x, y= 1 << m, 1 << m  # 最初の「位置」で偏りが0
for i in range(m + 1):  # 最小のx,yを見つけ出す。
    if d[H - 1][W - 1] & x or d[H - 1][W - 1] & y:
        print(i)
        quit()
    x<<=1
    y>>=1
```

```python
def dic_memory():
    for i in bak:
        for j in bak:
            if dic[i][j] == True:
                for k in bak:
                    if dic[j][k] == True or dic[k][j] == False:
                        dic[i][k] = True
                        dic[k][i] = False
            if dic[i][j] == False:
                for k in bak:
                    if dic[j][k] == False or dic[k][j] == True:
                        dic[i][k] = False
                        dic[k][i] = True


def heapify(unsorted, index, heap_size):
    largest = index
    left_index = 2 * index + 1
    right_index = 2 * index + 2
    dic_memory()
    dic_memory()
    dic_memory()
    if left_index < heap_size:
        if dic[unsorted[left_index]][unsorted[largest]] != '':
            lans = dic[unsorted[left_index]][unsorted[largest]]
        elif dic[unsorted[largest]][unsorted[left_index]] != '':
            lans = dic[unsorted[largest]][unsorted[left_index]]
        else:
            print('?', unsorted[left_index], unsorted[largest])
            lans = input()
        if lans == '>':
            dic[unsorted[left_index]][unsorted[largest]] = '>'
            dic[unsorted[largest]][unsorted[left_index]] = '<'
            largest = left_index
        else:
            dic[unsorted[left_index]][unsorted[largest]] = '<'
            dic[unsorted[largest]][unsorted[left_index]] = '>'
            
    if right_index < heap_size:
        if dic[unsorted[right_index]][unsorted[largest]] != '':
            rans = dic[unsorted[right_index]][unsorted[largest]]
        elif dic[unsorted[largest]][unsorted[right_index]] != '':
            rans = dic[unsorted[largest]][unsorted[right_index]]
        else:
            print('?', unsorted[right_index], unsorted[largest])
            rans = input()
        if rans == '>':
            dic[unsorted[right_index]][unsorted[largest]] = '>'
            dic[unsorted[largest]][unsorted[right_index]] = '<'
            largest = right_index
        else:
            dic[unsorted[right_index]][unsorted[largest]] = '<'
            dic[unsorted[largest]][unsorted[right_index]] = '>'
    if largest != index:
        unsorted[largest], unsorted[index] = unsorted[index], unsorted[largest]
        heapify(unsorted, largest, heap_size)


def heap_sort(unsorted):
    n = len(unsorted)
    for i in range(n // 2 - 1, -1, -1):
        heapify(unsorted, i, n)
    for i in range(n - 1, 0, -1):
        unsorted[0], unsorted[i] = unsorted[i], unsorted[0]
        heapify(unsorted, 0, i)
    return unsorted


N, Q = [int(item) for item in input().split()]
unsorted = []
for i in range(N):
    unsorted.append(chr(i+ord('A')))
bak = unsorted.copy()
s = ''
dic = {}
for i in bak:
    dic[i] = {}
    for j in bak:
        dic[i][j]=''

lst = heap_sort(unsorted)
for item in lst:
    s+=item
print('!', s)
```

```python
import numpy as np
from typing import List
class Solution:
    def minFlips(self, mat: List[List[int]]) -> int:
        mat_arr = np.array(mat)
        if not mat_arr.any():
            return 0
        m = len(mat)
        n = len(mat[0])
        c = 1 << 10
        arr = np.zeros((m + 2, n + 2), dtype=bool)
        arr[1:-1,1:-1] = np.array(mat)
        for i in range(1, 1 << m * n):
            tmp = arr.copy()
            ti = i
            tm = 0
            tn = 0
            while ti:
                if ti & 1:
                    tmp = self.press(tm, tn, tmp)
                tn += 1
                if tn == n:
                    tn = 0
                    tm += 1
                ti >>= 1
            if not tmp[1:-1,1:-1].any():
                c = min(c, self.count_one(i))
        if c == 1 << 10:
            return -1
        else:
            return c

    def press(self, i, j, arr):
        arr[i+1, j:j+3] = ~arr[i+1, j:j+3]
        arr[i,j+1], arr[i+2,j+1] = ~arr[i,j+1], ~arr[i+2,j+1]
        return arr

    def count_one(self, num):
        c = 0
        while num:
            num &= num - 1
            c += 1
        return c
```

```python
import sys
read = sys.stdin.buffer.read
readline = sys.stdin.buffer.readline
readlines = sys.stdin.buffer.readlines
import numpy as np

n, u, v = map(int, readline().split())
m = map(int,read().split())
data = list(zip(n, n))

graph = [[] for _ in range(n+1)]
gcopy = [[] for _ in range(n+1)]
for a,b in data:
    graph[a].append(b)
    graph[b].append(a)
    gcopy[a].append(b)
    gcopy[b].append(a)

def EulerTour(graph,start=1):
    V = len(graph)
    par = [0] * V
    depth = [0] * V; depth[start] = 0
    tour = [start]; st = [start]
    while st:
        x = st[-1]
        if not graph[x]:
            st.pop()
            tour.append(par[x])
            continue
        y = graph[x].pop()
        if y == par[x]:
            continue
        par[y] = x
        depth[y] = depth[x] + 1
        st.append(y)
        tour.append(y)
    return tour,depth

tour, depth = EulerTour(graph)

Ltour = len(tour)
tour_arr = np.array(tour)
depth_arr = np.array(depth)
tour_d = depth_arr[tour_arr]
idx = np.arange(len(depth))
idx[tour_arr] = np.arange(Ltour)

sp = np.empty((Ltour.bit_length(),Ltour),np.int32)
sp[0] = np.arange(Ltour)
for n in range(1,Ltour.bit_length()):
    prev,width = sp[n-1],1<<(n-1)
    x = prev[:-width]; y = prev[width:]
    condition = tour_d[x] > tour_d[y]
    sp[n] = prev
    sp[n,:-width][condition] = y[condition]

def LCA(A,B):
    AB = np.vstack([A,B]).T
    LR = idx[AB]
    LR.sort(axis=1)
    # [L,R] におけるRmQ
    L = LR[:,0]; R = LR[:,1]
    x = R-L
    n = np.zeros_like(x) # 2^n <= R-L
    for _ in range(20):
        x >>= 1
        n[x>0] += 1
    x = sp[n,L]; y = sp[n,R-(1<<n)+1]
    return np.where(tour_d[x] < tour_d[y], tour_arr[x], tour_arr[y])
w = LCA(u,v)
du = depth[u]
L = du + depth[v] - 2 * depth[w]
```

```python
import sys
read = sys.stdin.buffer.read
readline = sys.stdin.buffer.readline
readlines = sys.stdin.buffer.readlines
import numpy as np
n = int(readline())
odd = n // 2 + n % 2
goods = np.array([int(i) for i in readline().split()])
mingoods = np.min(goods)
minodd = np.min(goods[0::2])
_, *q = readlines()
res = 0
for i in q:
  num, *l = map(int, i.strip().split())
  if num == 1:
    t0 = l[0]
    tl = l[1]
    if goods[t0 - 1] >= tl:
      goods[t0 - 1] -= tl
      a = goods[t0 - 1]
      res += tl
      if mingoods > a:
        mingoods = a
      if t0%2==1 and minodd > a:
        minodd = a
  elif num == 2:
    tl = l[0]
    if minodd >= tl:
      goods[0::2] -= tl
      res += tl * odd
      minodd -= tl
      if mingoods > minodd:
        mingoods = minodd
  elif num == 3:
    tl = l[0]
    if mingoods >= tl:
      goods -= tl
      res += tl * n
      mingoods -= tl
print(res)
```

```python
from typing import List
class Solution:
    def shortestPath(self, grid: List[List[int]], k: int) -> int:
        m, n = len(grid), len(grid[0])
        start = m-1, n-1, 0
        queue = [([], start)]
        seen = {start}
        for steps, (i, j, cost) in queue:
#             if k >= i + j - 1:
#                 return steps + i + j
            for i, j in (i+1, j), (i-1, j), (i, j+1), (i, j-1):
                if m > i >= 0 <= j < n:
                    state = i, j, cost + grid[i][j]
                    if state not in seen:
                        steps.append([i, j])
                        queue.append((steps, state))
                        seen.add(state)
        return queue
```
